#### IMPORT LIBRARIES

```python
import cv2
from matplotlib import pyplot as plt
```

#### LOADING IMAGE

```python
image = cv2.imread("thanos.png")
color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (10,10))
plt.axis("off")
plt.imshow(color)
plt.title('Original Image')
plt.show
```

#### CONVERTING TO GREYSCALE

```python
grey_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('GreyScale Image')
plt.imshow(grey_img,cmap='gray')
plt.show()
```

#### INVERTING GREYSCALE

```python
inverted_img = 255 - grey_img
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('Inverted Image')
plt.imshow(inverted_img,cmap='gray')
plt.show()
```

#### BLURRING USING GAUSSIAN FUNCTION

```python
blurred = cv2.GaussianBlur(inverted_img, (21,21),0)
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('Blurred Image')
plt.imshow(blurred,cmap='gray')
plt.show()
```

#### INVERTING BLURRED IMAGE

```python
invert_blurred = 255 - blurred
pencil = cv2.divide(grey_img,invert_blurred,scale=256.0)
plt.figure(figsize = (10,10))
plt.axis('off')
plt.title('Pencil Image')
plt.imshow(pencil,cmap='gray')
plt.show()
```

#### COMPARING THE ORIGINAL IMAGE WITH THE NEW PENCIL SKETCH

```python
plt.figure(figsize = (10,10))
plt.subplot(1,2,1)
plt.imshow(color)
plt.axis('off')
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(pencil,cmap='gray')
plt.axis('off')
plt.title('Pencil Image')
```
